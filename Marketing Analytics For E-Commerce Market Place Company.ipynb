```python
# create all the Joins 
# Join the customer and order data 
cust_order=pd.merge(left=customers,right=orders,left_on='customer_id',right_on='customer_id',how='left')

# Join the orders and order_payments data 
orders_join_payments=pd.merge(left=orders,right=order_payments,left_on='order_id',right_on='order_id',how='right')

# Join the orders and order_review_ratings data 
orders_join_review_ratings=pd.merge(left=orders,right=order_review_ratings,left_on='order_id',right_on='order_id',how='right')

# Join the order and order_items data 
orders_join_items=pd.merge(left=orders,right=order_items,left_on='order_id',right_on='order_id',how='right')

# Join the order_items and products data 
orders_items_join_products=pd.merge(left=order_items,right=products,left_on='product_id',right_on='product_id',how='left')

# Join the sellers and order_items data 
sellers_join_order_items=pd.merge(left=order_items,right=sellers,left_on='seller_id',right_on='seller_id',how='left')

# Join the sellers and geo_locations data 
sellers_join_geolocation=pd.merge(left=sellers,right=geolocation,left_on='seller_zip_code_prefix',right_on='geolocation_zip_code_prefix',how='left')

# Join the customers and geo_locations data 
customers_join_geolocation=pd.merge(left=customers,right=geolocation,left_on='customer_zip_code_prefix',right_on='geolocation_zip_code_prefix',how='left')
```

## 3. Cross-Selling (Which products are selling together)
#### Hint: We need to find which of the top 10 combinations of products are selling together in each transaction. (combination of 2 or 3 buying together)

```python
from itertools import combinations
from collections import Counter
```

```python
items = orders_items_join_products[orders_items_join_products.order_id.duplicated(keep=False)]
items
```

```python
order_product_groups = items.groupby('order_id')['product_category_name'].agg(list).reset_index()

product_combinations = []

for product in order_product_groups.product_category_name:
    for comb in combinations(product,2):
        if len(set(comb)) == len(comb):
            product_combinations.append(comb)
    for comb in combinations(product,3):
        if len(set(comb)) == len(comb):
            product_combinations.append(comb)

product_combinations_count = Counter(product_combinations)

top_10_combinations = product_combinations_count.most_common(10)

print("Top 10 combinations of products sold together:")
for combination, count in top_10_combinations:
    print(f"Combination: {combination}, Count: {count}")
```
